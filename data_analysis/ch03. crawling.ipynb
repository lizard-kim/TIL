### beautiful soap 사용하기

```python
from bs4 import BeautifulSoup
```

```python
html = '''
<html>
  <head>
    <title>BeautifulSoup test</title>
  </head>
  <body>
    <div id='upper' class='test' custom='good'>
      <h3 title='Good Content Title'>Contents Title</h3>
      <p>Test contents</p>
    </div>
    <div id='lower' class='test' custom='nice'>
      <p>Test Test Test 1</p>
      <p>Test Test Test 2</p>
      <p>Test Test Test 3</p>
    </div>
  </body>
</html>'''
```

```python
soup = BeautifulSoup(html)
```

```python
soup.find('h3')
```

```python
soup.find('p') # 가장 첫번째
```

```python
soup.find('div', class_='test') # class는 _(언더바) 가 필요
```

```python
attrs = {
    'id': 'upper',
    'class': 'test',
}
soup.find(attrs=attrs)
```

```python
soup.find_all('div')
```

```python
tag = soup.find('h3')
print(tag)
tag.get_text()
```

```python
tag = soup.find('div', id='upper')
print(tag)
tag.get_text().strip()
```

```python
tag = soup.find('h3')
print(tag)
tag['title']
```

### requests + bs4

```python
url = 'https://news.jtbc.joins.com/html/026/NB11990026.html'
resp = requests.get(url)
```

```python
soup = BeautifulSoup(resp.content.decode('utf-8', 'replace')) # 한글 깨짐 방지
'''
r.text is the content of the response in Unicode, 
and r.content is the content of the response in bytes.
'''
title = soup.find('h3', id='jtbcBody')
print(title)
title.get_text()
```

```python
info = soup.find('span')
print(info)
print(info.find('span', class_='i_date'))
```
